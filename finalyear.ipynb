```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from numpy import dot
from numpy.linalg import norm
import numpy as np
from numpy.linalg import norm

# Load the GloVe embeddings
glove_embeddings = {}
with open('glove.6B.100d.txt', encoding='utf8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        embedding = np.array(values[1:], dtype='float32')
        glove_embeddings[word] = embedding

print(glove_embeddings['the'])
#print(shape(glove_embeddings['the']))
print(np.shape(glove_embeddings['the']))
```

```python
f1=open("/content/drive/MyDrive/sd3/similar/set1.txt","r")
f2=open("/content/drive/MyDrive/sd3/similar/set2.txt","r")
x=f1.readlines()
y=f2.readlines()
cos2=[]
for i in range(len(x)):
  print("\n",x[i],"\n",y[i])
  s1=x[i]
  t2=s1.split()
  s3=""
  for i in range(len(t2)):
    #print("first....",t2[i])
    c=0
    for j in range(len(t2[i])):
      if t2[i][j]=='-':
       c=1
       s4=t2[i][0:j]
       s5=t2[i][j+1:len(t2[i])]
       #print(s4,s5)
       break
    if c==0:
      s3=s3+t2[i]+" "
    else:
     s3=s3+s4+" "+s5+" "
  #print(s3)

  d1 = word_tokenize(s3.lower())
  s1=y[i]
  t2=s1.split()
  s3=""
  for i in range(len(t2)):
    #print("first....",t2[i])
    c=0
    for j in range(len(t2[i])):
      if t2[i][j]=='-':
       c=1
       s4=t2[i][0:j]
       s5=t2[i][j+1:len(t2[i])]
       #print(s4,s5)
       break
    if c==0:
      s3=s3+t2[i]+" "
    else:
     s3=s3+s4+" "+s5+" "
  #print(s3)

  d2= word_tokenize(s3.lower())
  sum1=0
  for j in range(len(d1)):
     d1[j]=d1[j].lower()
     #print(d1[j])
     if d1[j] not in glove_embeddings:
      pass
     else:
      sum1=sum1+glove_embeddings[d1[j]]
  sum1=sum1/len(d1)
  sum2=0
  for k in range(len(d2)):
     d2[k]=d2[k].lower()
     if d2[k] not in glove_embeddings:
      pass
     else:
      sum2=sum2+glove_embeddings[d2[k]]
  sum2=sum2/len(d2)
  cosine = np.dot(sum1,sum2)/(norm(sum1)*norm(sum2))
  print("Cosine Similarity:", cosine/2)
  cos2.append(cosine/2)

  #cos2 measures similarity using glove
```
